```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(30, 13))
mask = np.triu(np.ones_like(soal3.corr(), dtype=np.bool_))
heatmap = sns.heatmap(soal3.corr(), mask=mask, vmin=-1, vmax=1, annot=True, cmap='YlGn')
heatmap.set_title('Soal 3 (Korelasi Penyebab Kematian di Indonesia)', fontdict={'fontsize':18}, pad=16);
```

###Soal 4 Point 1

```python
soal4_1 = df2[df2['Area'].isin(['China', 'India', 'Indonesia', 'Thailand'])][['Area', 'Value']]
soal4_1 = soal4_1.groupby('Area')['Value'].sum().reset_index()
soal4_1
```

```python
#Bisa di Hover untuk cek detail nilai pada setiap Line Chart
import plotly.graph_objects as go

# create bar plot
fig = go.Figure()
fig.add_trace(go.Bar(x=soal4_1['Area'], y=soal4_1['Value'], 
                     marker_color=['#FFC300', '#FF5733', '#C70039', '#900C3F']))

# customize layout
fig.update_layout(title='Soal 4 Point 1 (Perbandingan Jumlah Produksi Beras)',
                  xaxis_title='Negara',
                  yaxis_title='Total Produksi',
                  width=800,
                  height=500,
                  title_x=0.5)

# show plot
fig.show()
```

###Soal 4 Point 2

```python
soal4_2 = df2[df2['Area'].isin(['China', 'India', 'Indonesia', 'Thailand'])][['Area', 'Year', 'Value']]
soal4_2 = soal4_2.sort_values('Year')
soal4_2
```

```python
table = pd.pivot_table(soal4_2, values='Value', index='Area', columns='Year')

# create a heatmap without the numbers
sns.heatmap(table, cmap='vlag', annot=False)

# set the title
plt.title('Heatmap')

# show the plot
plt.show()
```

```python
soal4_2_China     = df2.loc[df2['Area'] == 'China', ['Year', 'Value']]
soal4_2_Indonesia = df2.loc[df2['Area'] == 'Indonesia', ['Year', 'Value']]
soal4_2_India     = df2.loc[df2['Area'] == 'India', ['Year', 'Value']]
soal4_2_Thailand  = df2.loc[df2['Area'] == 'Thailand', ['Year', 'Value']]

soal4_2_China = soal4_2_China.sort_values('Year')
soal4_2_Indonesia = soal4_2_Indonesia.sort_values('Year')
soal4_2_India = soal4_2_India.sort_values('Year')
soal4_2_Thailand = soal4_2_Thailand.sort_values('Year')
```

```python
#Bisa di Hover untuk cek detail nilai pada setiap Line Chart

import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Scatter(y= soal4_2_China['Value'], x=soal4_2_China['Year'], mode="lines+markers", name="China"))
fig.add_trace(go.Scatter(y= soal4_2_Indonesia['Value'], x=soal4_2_Indonesia['Year'], mode="lines+markers", name="Indonesia"))
fig.add_trace(go.Scatter(y= soal4_2_India['Value'], x=soal4_2_India['Year'], mode="lines+markers", name="India"))
fig.add_trace(go.Scatter(y= soal4_2_Thailand['Value'], x=soal4_2_Thailand['Year'], mode="lines+markers", name="Thailand"))
```

Jika dilihat dari Heatmap dan Line Plot tidak ditemukan korelasi yang benar-benar positif dalam jumlah produksi beras antara China, India, Indonesia, dan Thailand. Hanya terlihat penurunan pada produksi beras di India tahun 2002 dan diikuti China pada tahun 2003, akan tetapi keduanya kembali naik pada tahun selanjutnya. Untuk Indonesia dan Thailand hampir memiliki kesamaan jumlah produksi di setiap tahunnya dan cenderung stabil jika dibandingkan India dan China walaupun produksinya tidak sebanyak China dan India.
